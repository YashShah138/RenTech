# Black Scholes Implementation

**The formula for a European call option is:**
- $ C(S, t) = S N(d_1) - K e^{-r(T - t)} N(d_2) $


**where:**
- $ d_1 = \frac{\ln\left(\frac{S}{K}\right) + \left(r + \frac{\sigma^2}{2}\right)(t)}{\sigma \sqrt{t}}, \quad \\$
- $ \\d_2 = d_1 - \sigma \sqrt{t}. $


**Definitions:**
- $ S: \text{Current stock price} $
- $ K: \text{Strike Price} $
- $ t: \text{Time to Maturity} $
- $ {\sigma}: \text{Volatility of the asset} $
- $ r: \text{Risk-Free Interest Rate} $
- $ N(x): \text{CDF of the normal distribution} $

```python
import math
from scipy.stats import norm

def blackScholes(s, k, t, r, sig):
    d1 = (math.log(s / k) + (r + 0.5 * sig ** 2) * t) / (sig * math.sqrt(t))
    d2 = d1 - sig*math.sqrt(t)

    return (s * norm.cdf(d1)) - (k * math.exp(-r * t) * norm.cdf(d2))


# Test implementation
s = 242.65      ## Current Stock Price
k = 275.5       ## Strike Price
t = 1            ## Time to Maturity (1 year = 1)
r = 0.0451      ## Risk-free Interest Rate (5% = 0.05)
sig = 0.1924    ## Volatility(20%)

call_price = blackScholes(s, k, t, r, sig)

print(call_price)
```

```python
import matplotlib.pyplot as plt

a = math.trunc(blackScholes(s, 200, t, r, sig))
b = math.trunc(blackScholes(s, 210, t, r, sig))
c = math.trunc(blackScholes(s, 220, t, r, sig))
d = math.trunc(blackScholes(s, 230, t, r, sig))
e = math.trunc(blackScholes(s, 240, t, r, sig))
f = math.trunc(blackScholes(s, 250, t, r, sig))
g = math.trunc(blackScholes(s, 260, t, r, sig))
h = math.trunc(blackScholes(s, 270, t, r, sig))
i = math.trunc(blackScholes(s, 280, t, r, sig))
j = math.trunc(blackScholes(s, 290, t, r, sig))
l = math.trunc(blackScholes(s, 300, t, r, sig))

prices = [a, b, c, d, e, f, g, h, i, j, l]
x = list(range(200, 310, 10))

plt.figure(figsize=(10, 8))

plt.axvline(x=240, color='red', linestyle='--', label='Current Stock Price')
plt.plot(x, prices, marker='o', label='Call Option Prices')

for i, value in enumerate(prices):
    plt.text(x[i], value + 1, str(value), ha='center', va='bottom', fontsize=9)

plt.xlabel('Strike Price')
plt.ylabel('Call Option Price')
plt.title('Price of Call Options as Strike Price Changes')
plt.legend()

plt.show
```
